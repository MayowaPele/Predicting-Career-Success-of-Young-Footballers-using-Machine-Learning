```python
import numpy as np
import pandas as pd
import streamlit as st
```

```python
# Load the datasets
df1 = pd.read_csv('players_15.csv')
df2 = pd.read_csv('players_16.csv')
df3 = pd.read_csv('players_17.csv')
df4 = pd.read_csv('players_18.csv')
df5 = pd.read_csv('players_19.csv')
df6 = pd.read_csv('players_20.csv')
df7 = pd.read_csv('players_21.csv')
df8 = pd.read_csv('players_22.csv')
```

```python
# Merge the datasets
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], ignore_index=True) # Merging all datasets into one
```

```python
# Check first few rows
df.head()
```

```python
# Count the number of columns
num_columns = df.shape[1]
num_columns
```

```python
# Check data information
df.info()
```

```python
# Delete unnecessary columns
columns_to_drop = ['player_url', 'long_name', 'real_face','player_face_url', 'club_logo_url', 'club_flag_url', 'nation_logo_url', 'nation_flag_url', 'dob',
                   'release_clause_eur', 'nation_jersey_number', 'nation_team_id', 'ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram',
                   'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'lcb', 'cb', 'rcb', 'rwb', 'rdm', 'lb', 'rb', 'gk']
df_new = df.drop(columns=columns_to_drop)
```

```python
# Check first few rows after dropping columns
df_new.head()
```

```python
# Delete few more columns 
columns_to_drop2 = ['sofifa_id', 'short_name', 'club_jersey_number', 'club_loaned_from', 'club_joined', 'club_contract_valid_until']
df_new2 = df_new.drop(columns=columns_to_drop2)
```

```python
df_new2.head()
```

```python
# Filter by ages between 17 and 21
df_filtered_young = df_new2[(df_new2['age'] >= 17) & (df_new2['age'] <= 21)]
```

```python
len(df_new2), len(df_filtered_young)
```

```python
# show columns with more than 50% missing values
missing_threshold = 0.5
missing_columns = df_filtered_young.columns[df_filtered_young.isnull().mean() > missing_threshold]
missing_columns
```

```python
# Remove columns with more than 50% missing values
df_filtered_young_cleaned = df_filtered_young.drop(columns=missing_columns)
```

```python
# remove some more unimapctful columns
columns_to_drop3 = ['ldm', 'cdm', 'international_reputation']

df_filtered_young_cleaned2 = df_filtered_young_cleaned.drop(columns=columns_to_drop3)
```

```python
# remove 'potential', '
```

```python
# Import important libraries for model building
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer 
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score 
```

```python
# Encode categorical columns 
label_enc = LabelEncoder() # Initialize the label encoder
for column in df_filtered_young_cleaned2.select_dtypes(include=['object']).columns:
    df_filtered_young_cleaned2[column] = label_enc.fit_transform(df_filtered_young_cleaned2[column].astype(str))
```

```python
# Get columns with missing values
missing_columns = df_filtered_young_cleaned2.columns[df_filtered_young_cleaned2.isnull().any()]
missing_columns
```

```python
# Get number of missing values in columns with any missing values
df_filtered_young_cleaned2[missing_columns].isnull().sum()
```

```python
# Fill missing values of pace, shooting, passing, dribbling, defending and physic with mean grouped by 'overall'
cols = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']
for col in cols:
    df_filtered_young_cleaned2[col] = df_filtered_young_cleaned2.groupby('overall')[col].transform(lambda x: x.fillna(x.mean()))
```

```python
df_filtered_young_cleaned2[missing_columns].isnull().sum()
```

```python
# delete rows of 'league level' and 'club_team_id' column with missing values
df_filtered_young_cleaned2 = df_filtered_young_cleaned2.dropna(subset=['league_level', 'club_team_id'])
```

```python
df_filtered_young_cleaned2.columns[df_filtered_young_cleaned2.isnull().any()]
```

```python
# Fill missing values of 'mentality composure' columns with mean grouped by 'overall'
df_filtered_young_cleaned2['mentality_composure'] = df_filtered_young_cleaned2.groupby('overall')['mentality_composure'].transform(lambda x: x.fillna(x.mean()))
```

```python
# Fill remaining missing values with mean
df_filtered_young_cleaned2.fillna(df_filtered_young_cleaned2.mean(), inplace= True)
```

```python
# Check if any columns still have missing values
df_filtered_young_cleaned2.columns[df_filtered_young_cleaned2.isnull().any()] # No columns should have missing values now
```

```python
# Define target and independent variables
X = df_filtered_young_cleaned2.drop(columns=['overall', 'potential', 'value_eur', 'wage_eur']) # Independent variables
Y = df_filtered_young_cleaned2['overall'] # Target variable
```

```python
# Select top 15 features using f_regression
selector = SelectKBest(score_func=f_regression, k=15)
selector.fit(X, Y)
```

```python
# Rank the selected features by score 
feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': selector.scores_}).sort_values(by='Score', ascending=False)
```

```python
# Select the top 15 features
top_15_ranked = feature_scores.head(15)
selected_features = top_15_ranked['Feature'].values
X = X[selected_features]
```

```python
# Train and test split 
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Initialize models
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(max_depth=10, random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=50, random_state=42, max_depth=20)
}
```

```python
# Train and evaluate each model
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    print(f"\n{name}")
    print(f" Selected Fearures: {list(selected_features)}")
    print(f" Mean Squared Error (MSE): {mean_squared_error(y_test, y_pred)}")
    print(f" R-squared: {r2_score(y_test, y_pred):.2f}")
          
```
